# DIFFERENCE OF ARRAY & LIST

```python
import numpy as np
st= 'Hello Everyone!'
a=np.array(list(st))
b= st.split(' ')
print(a) #This is array
print(b) #This is List

c=np.array(b) #This is also array because it has no comma between values
print(c)
```

# IMAGES PROCESSING AS A NUMPY ARRAY

```python
import matplotlib.pylab as plt

def plti(im,h=8,**kwargs):
    y= im.shape[0]
    x=im.shape[1]
    w = (y/x)*h
    plt.figure(figsize=(w,h))
    plt.imshow(im,interpolation="none",**kwargs)
    plt.axis('off')
  
%matplotlib inline
im = plt.imread("air.jpg")
print(im.shape)
plti(im)
plti(im[:100,:100,:])
```
